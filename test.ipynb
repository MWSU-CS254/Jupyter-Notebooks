```java
public class Student
{
    private String name;
    private int score;

    public Student(String name, int score)
    {
        this.name = name;
        this.score = score;
    }

    public String getName()
    {
        return name;
    }

    public int getScore()
    {
        return score;
    }

    public boolean isPassing()
    {
        return score >= 70;
    }

    @Override
    public String toString()
    {
        return name + " scored " + score + ". Passing: " + isPassing();
    }
}
```

```java
Student student = new Student("Ada", 94);
System.out.println(student);
```

```java
import java.util.ArrayList;

ArrayList<Student> students = new ArrayList<>();
students.add(new Student("Ada", 94));
students.add(new Student("Grace", 88));
students.add(new Student("Alan", 67));

for (Student s : students)
{
    System.out.println(s);
}
```

```java
public int averageScore(ArrayList<Student> students)
{
    int total = 0;

    for (Student student : students)
    {
        total += student.getScore();
    }

    return total / students.size();
}

System.out.println("Average score: " + averageScore(students));
```
